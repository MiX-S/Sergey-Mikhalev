# Задача 2. Точки и отрезки

Дано `N` отрезков на числовой прямой и `M` точек на этой же прямой. Для каждой из данных точек определите, скольким отрезкам она принадлежит. Точка $x$ считается принадлежащей отрезку с концами $a$ и $b$, если выполняется двойное неравенство $\min(a, b) \le x \le \max(a, b)$.

На вход подается число `N`, затем `N` отрезков: каждый в отдельной строке. И в конце на вход подается `M` точекб разделенных пробелами. Границы отрезков и точки - вещественные числа.

На выходе должно быть `M` целых чисел, разделенных пробелами: количества отрезков, которым принадлежат точки. **Количества отрезков должны быть выведены в том же порядке, в котором точки были поданы на вход.**

### Примеры работы программы

| <font size=3>Входные данные</font> | <font size=3>Выходные данные</font> |
| :---: | :---: |
| <font size=3>2<br>1 2<br>2 3<br>0 1 2 3 4</font> | <font size=3>0 1 2 1 0</font> |
| <font size=3>3<br>1 10<br>-1 3<br>7 12<br>-6 -1 0 1 3 4 7 10 12 13</font> | <font size=3>0 1 1 2 2 1 2 2 1 0</font> |
| <font size=3>3<br>-1 10<br>-2 4<br>6 20<br>-2 4 1 -7 40 2.5 -0.3</font> | <font size=3>1 2 2 0 0 2 2</font> |

### Критерии

1. Если задача правильно решена перебором с помощью алгоритма $O(N^2)$, то начисляется 1 балл.

2. Если задача оптимально считает отрезки с эффективностью $O(N \log N)$, то начисляются 2,5 балла.

3. Правильные решения с эффективностью $O(N \log N)$, имеющие недочеты, оцениваются в 2 балла.

4. Дополнительно 0,5 балла можно получить, если запомнить порядок следования точек во входных данных и использовать его для расположения в правильном порядке  количеств отрезков, в которые  попали точки. Восстановление порядка должно работать также быстро, как "легкая сортировка" из `train1.ipynb`.

````python
def easy_sort(L, pos_idx):
    N = len(L)
    counter = 0
    for e_idx in range(N):
        while e_idx != L[e_idx][pos_idx]:
            j = L[e_idx][pos_idx]
            L[e_idx], L[j] = L[j], L[e_idx]
            counter += 1
        if counter >= N:
            break
        counter += 1

N = int(input())

left = []
right = []

for _ in range(N):
    l, r = map(float, input().split())
    left.append(l)
    right.append(r)

dots = list(map(float, input().split()))
M = len(dots)

dots_with_pos = list(zip(range(M), dots))

left.sort()
right.sort()
dots_with_pos.sort(key=lambda x: x[1])

positions, sorted_dots = zip(*dots_with_pos)

nsec = []
nsec_tmp = 0
l_idx, r_idx = 0, 0
for dot in sorted_dots:
    while l_idx < N and dot >= left[l_idx]:
        l_idx += 1
        nsec_tmp += 1
    while r_idx < N and dot > right[r_idx]:
        r_idx += 1
        nsec_tmp -= 1
    nsec.append(nsec_tmp)
        
nsec_with_pos = list(zip(positions, nsec))

easy_sort(nsec_with_pos, 0)

_, nsec_sorted = zip(*nsec_with_pos)
print(' '.join(map(str, nsec_sorted)))
````

# Задача 3. Аттракцион

На протяжении многих лет в Байтландии существует парк развлечений “Funny byte”, в котором представлено много различных аттракционов: колесо вычислений, веселые горки с трассами в форме интегралов, равнобедренная ромашка и многие другие.

В последние годы популярность “Funny byte” стала неуклонно падать. В целях привлечения посетителей руководство парка решило открыть новый аттракцион, который представляет собой сложный механизм.

Кресла аттракциона расположены в `N` рядов по `M` кресел в каждом. То есть каждое кресло характеризуется номером ряда и номером колонки, в котором оно находится. Ряды нумеруются последовательно сверху вниз начиная с единицы, колонки нумеруются слева направо начиная с единицы. Каждое кресло имеет свой уникальный номер. Кресло, находящееся в `i`-м ряду и в `j`-ой колонке, имеет номер `(i-1)*M + j`.

После посадки отдыхающих, кресла поднимают вверх над землей. И начинается веселье! Механизм аттракциона случайным образом производит некоторое количество операций. Под одной операцией понимается взаимная перестановка двух рядов либо двух колонок. При взаимной перестановке двух рядов или колонок каждое кресло в ряду или колонке заменятся на соответствующее ему кресло в другом ряду или колонке.

И вот аттракцион завершил свою работу. Но есть одна трудность! Кресла надо вернуть в начальное положение при помощи таких же операций. Как количество, так и сами операции не обязательно должны быть идентичны тем, которые производил аттракцион во время сеанса. Руководство парка решило, что вернуть кресла в начальное положение необходимо не более чем за `(N + M)` операций.

Такая задача оказалась не по силам разработчикам механизма. Помогите им! От вас требуется разработать программу, позволяющую вернуть кресла в начальное положение. Гарантируется, что решение существует.

На вход подается число `N`, а затем `N` строк из `M` целых чисел, разделенных пробелами, в каждой. Каждая `k`-я поданная на вход строка содержит расположение кресел в `k`-м ряде после завершения работы аттракциона. 

Программа должна распечатать перестановки в аттракционе, которые требуется выполнить, чтобы привести его в порядок. Каждая перестановка печатается в отдельной строке и состоит из трех чисел, разделенных пробелами.

1. Если первое число равно 0, то переставляются ряды, а если 1, то переставляются колонки.

2. Второе и третье числа - номера переставляемых колонок или рядов. **Сначала печатается меньший номер, затем больший.**

**Если есть несколько решений, можно привести любое из них.**

# Задача 4. Устойчивая сортировка выбором

Реализуйте устойчивую версию сортировки выбором `stable_selection_sort()`. Функция должна принимать на вход список чисел и сортировать его на месте, то есть переставлять элементы оригинального списка. Функция не должна ничего возвращать.

Запрещается использовать готовые решения, найденные в сети интернет.

### Критерии

За рабочее решение начисляются 1,5 балла

````python
def index_of_min(L):
    min_ = float('inf')
    for i, el in enumerate(L):
        if el < min_:
            i_of_min  = i
            min_ = el
    return i_of_min


def stable_selection_sort(L):
    N = len(L)
    for i in range(N-1):
        i_of_min = index_of_min(L[i:]) + i
        tmp = L[i_of_min]
        for k in range(i_of_min, i, -1):
            L[k] = L[k-1]
        L[i] = tmp
        

import random
N = 10
L = random.sample(list(range(N)), N)
stable_selection_sort(L)
````

````python
def shell_sort(L):
    N = len(L)
    d = N // 2
    while d > 0:
        for i in range(d, N):
            tmp = L[i]
            j = i
            while j >= d and tmp < L[j-d]:
                L[j] = L[j-d]
                j -= d
            L[j] = tmp
        d //= 2
````

````python
def get_pratt_seq(N):
    n = 1
    seq = []
    while n <= N:
        nn = n
        while nn <= N:
            seq.append(nn)
            nn *= 3
        n *= 2
    seq.sort()
    return seq          
````

````python
def pratt_sort(L):
    N = len(L)
    pratt_seq = get_pratt_seq(N)[::-1]
    for d in pratt_seq:
        for i in range(d, N):
            tmp = L[i]
            j = i
            while j >= d and tmp < L[j-d]:
                L[j] = L[j-d]
                j -= d
            L[j] = tmp
````

````python
def comb_sort(L):
    N = len(L)
    factor = 1.247
    d = int(N / factor)
    correct = False
    while not correct:
        if d == 1:
            correct = True
        for i in range(d, N):
            if L[i] < L[i-d]:
                correct = False
                L[i], L[i-d] = L[i-d], L[i]
        d = int(d / factor)
        if d < 1:
            d = 1
````

````python

def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:

      for startposition in range(sublistcount):
        gapInsertionSort(alist,startposition,sublistcount)

      sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):

        currentvalue = alist[i]
        position = i

        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap

        alist[position]=currentvalue
````

````python
import timeit
import math
import matplotlib.pyplot as plt
start = 8
factor = 2
N_max = 2**16
num_repeats = 5
N_list = []
T_shell_list = []
T_shell2_list = []
T_pratt_list = []
T_comb_list = []


setup = """
import random
L=list(range(N))
random.shuffle(L)
"""
N = start
while N <= N_max:
    T_shell = timeit.timeit(
        stmt="a = L[:];shell_sort(a)",
        setup=setup,
        number=num_repeats,
        globals=dict(N=N, shell_sort=shellSort, gapInsertionSort=gapInsertionSort)
    )
    T_shell2 = timeit.timeit(
        stmt="a = L[:];shellSort(a)",
        setup=setup,
        number=num_repeats,
        globals=dict(N=N, shellSort=shellSort, gapInsertionSort=gapInsertionSort)
    )
    T_comb = timeit.timeit(
        stmt="a = L[:];comb_sort(a)",
        setup=setup,
        number=num_repeats,
        globals=dict(N=N, comb_sort=comb_sort)
    )
    T_pratt = timeit.timeit(
        stmt="a = L[:];pratt_sort(a)",
        setup=setup,
        number=num_repeats,
        globals=dict(N=N, pratt_sort=pratt_sort, get_pratt_seq=get_pratt_seq)
    )
    T_copy = timeit.timeit(
        stmt="a = L[:]",
        setup=setup,
        number=num_repeats,
        globals=dict(N=N)
    )
    N_list.append(N)
    T_shell_list.append((T_shell - T_copy) / num_repeats)
    T_shell2_list.append((T_shell2 - T_copy) / num_repeats)
    T_pratt_list.append((T_pratt - T_copy) / num_repeats)
    T_comb_list.append((T_comb - T_copy) / num_repeats)
    print("{} length done".format(N))
    N *= factor
    
plt.plot(N_list, T_shell_list, label='shell')
plt.plot(N_list, T_shell2_list, label='shell2')
plt.plot(N_list, T_pratt_list, label='pratt')
plt.plot(N_list, T_comb_list, label='comb')
plt.xlabel(r"${N}$", fontsize=18)
plt.ylabel(r"$T, seconds$", fontsize=18)
plt.xscale('linear')
plt.yscale('linear')
plt.legend(** dict(
                bbox_to_anchor=(.02, .98),
                loc=2,
            ))
# plt.title('timsort: test time estimate $O(N \cdot log N)$', fontsize=16)
plt.show()
````

# Задача 6. $k$-й по величине элемент

Найти $k$-е по величине значение в массиве из $N$ элементов (не сортируя массив полностью). Можно менять исходный массив.

Формат входных и выходных данных не имеет принципиального значения. Решением будет функция `quick_search()`, принимающая список и позицию элемента в отсортированном массиве.

### Критерии

За рабочее решение со средней эффективностью $O(N)$ начисляются 3 балла.

````python
def dual_pivot_qsort(A, lo, hi):
    if hi - lo >= 1:
        bounds = partition(A, lo, hi)
        if bounds is not None:
            dual_pivot_qsort(A, lo, bounds[0]-1)
            dual_pivot_qsort(A, bounds[0]+1, bounds[1]-1)
            dual_pivot_qsort(A, bounds[1]+1, hi)
        
        
def find_and_place_pivots(A, lo, hi):
    if A[lo] > A[hi]:
        A[lo], A[hi] = A[hi], A[lo]
    elif A[lo] == A[hi]:
        i = lo + 1
        while i < hi and A[i] == A[i-1]:
            i += 1
        if A[i] < A[lo]:
            A[lo], A[i] = A[i], A[lo]
        elif A[i] > A[lo]:
            A[hi], A[i] = A[i], A[hi]
    return A[lo], A[hi]
        
    
def partition(A, lo, hi):
    piv1, piv2 = find_and_place_pivots(A, lo, hi)
    if piv1 == piv2:
        return
    left = lo + 1
    mid = lo + 1
    right = hi - 1
    while mid <= right:
        if A[mid] < piv1:
            A[mid], A[left] = A[left], A[mid]
            left += 1
            mid += 1
        elif A[mid] > piv2:
            while A[right] > piv2 and mid <= right:
                right -= 1
            if mid < right:
                A[mid], A[right] = A[right], A[mid]
            if A[right] > piv2:
                right -= 1
        else:
            mid += 1
    A[lo], A[left-1] = A[left-1], A[lo]
    A[hi], A[right+1] = A[right+1], A[hi]
    return left-1, right+1
        
A = [2, 1, 7, 5, 4]
# A = [3, 2, 3, 7, 1, 8, 6, 6, 6, 6, -1, 5]
dual_pivot_qsort(A, 0, len(A)-1)
print(' '.join(map(str,A)))
````

````python
def quick_search(A, lo, hi, k):
    if hi - lo >= 1:
        bounds_and_pivots = partition(A, lo, hi)
        if bounds_and_pivots is None:
            return A[lo]
        else:
            if bounds_and_pivots[0] == k:
                return bounds_and_pivots[2]
            elif bounds_and_pivots[1] == k:
                return bounds_and_pivots[3]
            elif k < bounds_and_pivots[0]:
                return quick_search(A, lo, bounds_and_pivots[0], k)
            elif bounds_and_pivots[0] < k < bounds_and_pivots[1]:
                return quick_search(A, bounds_and_pivots[0]+1, bounds_and_pivots[1]-1, k)
            else:
                return quick_search(A, bounds_and_pivots[1], hi, k)
    elif hi == lo:
        return A[lo]
        
        
def find_and_place_pivots(A, lo, hi):
    if A[lo] > A[hi]:
        A[lo], A[hi] = A[hi], A[lo]
    elif A[lo] == A[hi]:
        i = lo + 1
        while i < hi and A[i] == A[i-1]:
            i += 1
        if A[i] < A[lo]:
            A[lo], A[i] = A[i], A[lo]
        elif A[i] > A[lo]:
            A[hi], A[i] = A[i], A[hi]
    return A[lo], A[hi]
        
    
def partition(A, lo, hi):
    piv1, piv2 = find_and_place_pivots(A, lo, hi)
    if piv1 == piv2:
        return
    left = lo + 1
    mid = lo + 1
    right = hi - 1
    while mid <= right:
        if A[mid] < piv1:
            A[mid], A[left] = A[left], A[mid]
            left += 1
            mid += 1
        elif A[mid] > piv2:
            while A[right] > piv2 and mid <= right:
                right -= 1
            if mid < right:
                A[mid], A[right] = A[right], A[mid]
            if A[right] > piv2:
                right -= 1
        else:
            mid += 1
    A[lo], A[left-1] = A[left-1], A[lo]
    A[hi], A[right+1] = A[right+1], A[hi]
    return left-1, right+1, piv1, piv2
   
A = [3, 2, 3, 7, 1, 8, 6, 6, 6, 6, -1, 5]
for i in range(len(A)):
    A = [3, 2, 3, 7, 1, 8, 6, 6, 6, 6, -1, 5]
    print(quick_search(A, 0, 11, i))
````

# Задача 7. Блочная сортировка

`N` точек равномерно распределены внутри единичного круга с центром в начале координат. Отсортируйте их по тому, насколько они близки к началу координат. Испольуйте блочную сортировку.

### Критерии

За рабочее решение начисляются 2 балла.

Вычисление $r_1$

$$\pi r_1^2 = \frac{\pi}{N}$$

$$r_1 = \sqrt{\frac{1}{N}}$$

Вычисление $r_2$

$$\pi r_2^2 - \pi r_1^2 = \frac{\pi}{N}$$

$$r_2^2 = \frac{2}{N}$$

$$r_2 = \sqrt{\frac{2}{N}}$$

По индукции получаем, что

$$\pi r_n^2 - \pi r_{n-1}^2 = \frac{\pi}{N}$$

$$\pi r_n^2 - \pi \frac{n-1}{N} = \frac{\pi}{N}$$

$$r_n = \sqrt{\frac{n}{N}}$$

````python
def gt(p1, p2):
    return p1[0]**2 + p1[1]**2 > p2[0]**2 + p2[1]**2


def r2(p):
    return p[0]**2 + p[1]**2


def insertion_sort(L):
    n = len(L)
    for i in range(1, n):
        tmp = L[i]
        j = i - 1
        while gt(L[j], tmp) and j >= 0:
            L[j+1] = L[j]
            j -= 1
        L[j+1] = tmp
        
def bucket_sort(L, k):
    n = len(L)
    buckets = [[] for _ in range(k)]
    step = 1 / k
    for e in L:
        i = 0
        d = e[0]**2 + e[1]**2
        while d >= i * step:
            i += 1
        buckets[i-1].append(e)
    pointer = 0
    for bucket in buckets:
        insertion_sort(bucket)
        ln = len(bucket)
        L[pointer:pointer+ln] = bucket
        pointer += ln
````

````python
import random
N = 10
count = 0
points = []
while count < N:
    p = (random.uniform(-1, 1), random.uniform(-1, 1))
    if r2(p) <= 1:
        points.append(p)
        count += 1
        
print(points)
print(list(map(r2, points)))
bucket_sort(points, 3)
print(points)
print(list(map(r2, points)))
````

Напишите функцию для сортировки слов английского алфавита с помощью алгоритма LSD.

### Критерии

За рабочее решение начисляются 2,5 балла.

````python
import math
import string


def get_voc(L):
    voc = ''
    for w in  L:
        for c in w:
            if c not in voc:
                voc += c
    return ''.join(sorted(voc))


def count_sort(L, rad, voc):
    n = len(L)
    buf = [0]*n
    num_buckets = len(voc)
    count = [0]*num_buckets
    for e in L:
        try:
            count[voc.find(e[rad])] += 1
        except IndexError:
            raise
    for i in range(1, num_buckets):
        count[i] += count[i-1]
    for i in range(n-1, -1, -1):
        c_idx = voc.find(L[i][rad])
        count[c_idx] -= 1
        buf[count[c_idx]] = L[i]
    L[:] = buf


def radix_sort(L):
    r_max = len(max(L, key=len))
    voc = '0' + get_voc(L)
    for i in range(len(L)):
        L[i] = ('{:0<%s}' % r_max).format(L[i])
    for r in range(r_max):
        count_sort(L, r_max-r-1, voc)
    for i in range(len(L)):
        L[i] = L[i].rstrip('0')
        
import random
# L = list(range(150)) + list(range(100, 250))
L = [
        ''.join(
            random.sample(string.ascii_letters, random.randint(1, 10))
        )
        for _ in range(100)
    ]

print('\n'.join(map(str, L)), end='\n'*2)
radix_sort(L)
print(L)

'ab'[4:6]

# Задача 9. Удаление среза списка

Допишите в реализацию двусвязного списка метод, позволяющий удалять срезы с шагом 1 с помощью команды
```python
del my_list[i:j]
```

Случай, когда границы среза выходят за границы списка не рассматривать.

### Критерии

3 балла начисляются при условии, что **все** связи в списке корректно модифицируются, а `self.head` и `self.tail` указывают на действительные голову и хвост списка. 

>Для проверки обратных связей рекомендуется добавить и отладить метод `print_reverse()`, который будет печатать элементы списка, начиная с хвоста.

class Node:
    def __init__(self, data, next_node=None, prev_node=None):
        self.data = data
        self.next_node = next_node
        self.prev_node = prev_node

    def get_data(self):
        return self.data

    def set_data(self, val):
        self.data = val

    def get_next_node(self):
        return self.next_node
    
    def get_prev_node(self):
        return self.prev_node
    
    def set_next_node(self, node):
        self.next_node = node
    
    def set_prev_node(self, node):
        self.prev_node = node


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size(self):
        return self.size

    def add_node_to_head(self, data):
        new_node = Node(data, self.head, None)
        if self.head is None:
            self.tail = new_node
        else:
            self.head.set_prev_node(new_node)
        self.head = new_node
        self.size += 1
       
    def print_list(self):
        curr = self.head
        while curr:
            print(curr.data)
            curr = curr.get_next_node()
            
    def find_node(self, i):
        counter = 0
        node = self.head
        while counter < i:
            node = node.get_next_node()
            counter += 1
        return node
            
    def __delitem__(self, i):
        start, stop, _ = i.indices(self.size)
        start_node = self.find_node(start)
        first_node_after_slice = self.find_node(stop)
        last_node_before_slice = start_node.get_prev_node()
        if last_node_before_slice is not None:
            if first_node_after_slice is not None:
                last_node_before_slice.set_next_node(first_node_after_slice)
                first_node_after_slice.set_prev_node(last_node_before_slice)
            else:
                last_node_before_slice.set_next_node(None)
                self.tail = last_node_before_slice
        else:
            if first_node_after_slice is not None:
                self.head = first_node_after_slice
                first_node_after_slice.set_prev_node(None)
            else:
                self.head = None
                self.tail = None
        self.size -= stop - start

my_list = DoublyLinkedList()
print("Inserting")
for i in range(10):
    my_list.add_node_to_head(i)

print("Printing")
my_list.print_list()

del my_list[4:7]
print("Printing")
my_list.print_list()
print("Size")
print(my_list.get_size())

````
